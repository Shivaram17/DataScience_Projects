```python
import pandas as pd

dtypes = {
          'first_name': 'category',
          'gender': 'category',
          'type': 'category',
          'state': 'category',
          'party': 'category'
}
df = pd.read_csv('legislators-historical.csv', dtype=dtypes, parse_dates = ['birthday'], usecols = list(dtypes) + ['birthday', 'last_name'])
df.head()
```

```python
# basic groupby
n_by_state = df.groupby(['state', 'gender'])['last_name'].count()
n_by_state.head()
# type
type(n_by_state)
```

```python
# basic groupby with sub dataframe and also sort is set to False in groupby
n_by_state = df.groupby(['state', 'gender'], as_index = False, sort = False)['last_name'].count()
n_by_state.head()
# type
# type(n_by_state)
```
